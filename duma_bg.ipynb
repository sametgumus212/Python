```python
def adresscreator(url,all_link):
  adres_array=[]
  txt=""
  s=1
  for i in all_link:
    
    i=int(i)
    for j in range(1,i,1):
      txt=("{}{}{}{}".format(url,s,"&page=",j))
      adres_array.append(txt)
      print(txt)
    s+=1
  return adres_array

```

```python

def all_links(linker):
  array=[]
  for i in range(1,8,1):
    link=("{}{}".format(linker,i))
    #print(link)
    count=pagecountfinder(link)
    array.append(count)
    #print(count)
  return array
    
```

```python
def pagecountfinder(baselink):
  ar=[]
  req = requests.get(baselink)
  soup = BeautifulSoup(req.content, "lxml")
  title=soup.find("ul",attrs={"class":"pagination"}).findAll('a')
  txt=""
  for i in title:
    txt=i.getText()  
    ar.append(txt) 
  return(ar[-2])
```

```python
def pagelinkcreator(link):    
  dblink=[]
  req = requests.get(link)
  soup = BeautifulSoup(req.content, "lxml")
  title=soup.find_all("h2",attrs={"class":"cap"})
  title2=soup.find_all("h3",attrs={"class":"cap"})  
  k=""
  counter=0
  for i in title:
    b=i.a.get("href")
    k=b
    if (k.startswith(".")):
      pass
    else:
      #dblink.append("{}{}".format(link,k)) 
      dblink.append("{}{}".format("https://duma.bg/",k)) 
  for i in title2:  
    c=i.a.get("href")
    #dblink.append("{}{}".format(link,c))
    dblink.append("{}{}".format("https://duma.bg/",c))
    counter+=1
    if counter==10:
      break
  return dblink
```

```python
dblinks=[]
for i in db_adres:
  dblinks.append(pagelinkcreator(i))
```

```python
def creator(link):
  inner_array=[]
  atxt=""
  inner_array.append(link)    
  r = requests.get(link)   
  soup = BeautifulSoup(r.content, 'html5lib')
  dat=soup.find("p",attrs={"class":"date"}).time.get("datetime")
  inner_array.append(dat)
  title=soup.find("h1",attrs={"class":"cap"}).getText()
  inner_array.append(title)      
  table = soup.findAll("div", attrs = {"class":"news_dtl_body"})
  for i in table:
    ad=i.findAll("p",attrs={"class":""})
    for i in ad:
      txt=i.getText()
      if(len(txt)>40):
        atxt+=txt
  inner_array.append(atxt) 
  # ps=""
  # with open("cput.txt", 'a') as file: 
  #   #ps="{}{}{}{}{}{}{}".format(pdd[0],";",pdd[1],";",pdd[2],";",pdd[3])
  #   ps="{}{}{}{}{}{}{}".format(array[0],";",array[1],";",array[2],";",array[3])
  #   file.write(ps+'\n')
  #   file.truncate()
  return inner_array
```

```python
def work(links):
  arrays=[]
  for link in links:
    print(link)
    pdd=creator(link)      
    arrays.append(pdd)    
    ps=""
    with open("all_duma_bg_output.txt", 'a') as file: 
      ps="{}{}{}{}{}{}{}".format(pdd[0],";",pdd[1],";",pdd[2],";",pdd[3])
      file.write(ps+'\n')
      file.truncate()
  print(arrays)
```
