```python
def is_exist(link,c):
  #print(link)
  count=c
  r = requests.get(link)   
  soup = BeautifulSoup(r.content, 'html5lib')
  art=soup.findAll("div",attrs={"class":"article"}) 
  # print(count)
  if len(art)>1:
    if count<10:
      # print(count)
      count+=1
      new_link="{}{}".format(link[:-1],count)
      return is_exist(new_link,count)
    elif count<100:
      # print(count)
      count+=1
      new_link="{}{}".format(link[:-2],count)
      return is_exist(new_link,count)
    elif count<1000:
      # print(count)
      count+=1
      new_link="{}{}".format(link[:-3],count)
      return is_exist(new_link,count)
    # return True
  else:        
     return count-1
    
  
  # print(art)
```
